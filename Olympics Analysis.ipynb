### Graph Of Total Madels Over The Year Of Enterd Country

```python
def country_lineplot(country):
    nation_df = summer_athe_df[summer_athe_df['region'] == country]
    nation_year = nation_df.groupby('Year').sum().reset_index()
    nation_year['Total'] = nation_year['Gold'] + nation_year['Silver'] + nation_year['Bronze']
    plt.figure(figsize=(15,7))
    plt.plot(nation_year['Year'], nation_year['Total'])
    plt.xlabel('Year')
    plt.ylabel('Total Medals')
    plt.title('Total Medals Over The Year',fontdict={'fontsize': 20})
    plt.legend()
    return plt.legend()
```

```python
country_lineplot(country)
```

### Graph Of Different Medals Over Years Of Enterd Country

```python
def country_lineplot2(country):
    nation_df = summer_athe_df[summer_athe_df['region'] == country]
    nation_year = nation_df.groupby('Year').sum().reset_index()
    plt.figure(figsize=(15,7))
    plt.plot(nation_year['Year'], nation_year['Gold'], label='Gold Medals')
    plt.plot(nation_year['Year'], nation_year['Silver'], label='silver Medals')
    plt.plot(nation_year['Year'], nation_year['Bronze'], label='Bronze Medals')
    plt.xlabel('Year')
    plt.ylabel(' Medals')
    plt.title('different Medals Over The Year',fontdict={'fontsize': 20})
    plt.legend()
    return plt.title('different Medals Over The Year',fontdict={'fontsize': 20})
```

```python
country_lineplot2(country)
```

### Yearwise Medals Of Enterd Country

```python
def yearwise_medals_of_country(country):
    nation_df = summer_athe_df[summer_athe_df['region'] == country]
    a = nation_df.groupby('Year').sum()[['Gold','Silver','Bronze']]
    a['Total'] = a['Gold'] + a['Silver'] + a['Bronze']
    a.reset_index()
    return a.reset_index()
```

```python
yearwise_medals_of_country(country)
```

### Overall SportsWise Medals Of Enterd Country 

```python
def overall_total_medals_of_country(country):
    nation_df = summer_athe_df[summer_athe_df['region'] == country]
    a = nation_df.groupby('Sport').sum()[['Gold','Silver','Bronze']]
    a['Total'] = a['Gold'] + a['Silver'] + a['Bronze']
    a = a.sort_values('Gold', ascending=False).reset_index()
    return a
```

```python
overall_total_medals_of_country(country)
```

### Overall top 15 Player Of Enterd Country

```python
def top_15_player_of_country(country):
    nation_df = summer_athe_df[summer_athe_df['region'] == country]
    b = nation_df.groupby('Name').sum()[['Gold','Silver','Bronze']]
    b['Total'] = b['Gold'] + b['Silver'] + b['Bronze']
    b= b.sort_values('Total',ascending=False)
    b = b.reset_index()
    top_players = b.loc[0:14, :]
    top_players
    return top_players
```

```python
top_15_player_of_country(country)
```

### Male vs Female Medals Over Years Of Enterd Country 

```python
def male_vs_female_medals_of_country(country):
    nation_df = summer_athe_df[summer_athe_df['region'] == country]
    c= nation_df.groupby(['Year','Sex']).sum().reset_index()
    c['Total Medals'] = c['Gold'] + c['Silver'] + c['Bronze']
    plt.figure(figsize=(15,7))
    sns.scatterplot(x=c['Year'], y=c['Total Medals'], hue=c['Sex'], size=(25))
    plt.show()
    return plt.show()
```

```python
male_vs_female_medals_of_country(country)
```

### Participation Of Male, Female in Sport Of Enterd Country

```python
def participant_male_female(country):
    nation_df = summer_athe_df[summer_athe_df['region'] == country]
    plt.figure(figsize=(25,20))
    sns.scatterplot(x=nation_df['Year'], y=nation_df['Sport'], hue=nation_df['Sex'] )
    plt.title('Participation Of Male, Female in Sport',fontdict={'fontsize': 30})
    plt.show()
    return plt.show()
```

```python
participant_male_female(country)
```

### List Of Medals Eventwise Of Enterd Country

```python
def eventwise_medals(country):
    nation_df = summer_athe_df[summer_athe_df['region'] == country]
    c= nation_df.groupby(['Event']).sum()[['Gold', 'Silver', 'Bronze']].sort_values('Gold', ascending= False).reset_index()
    return c
```

```python
eventwise_medals(country)
```

### Event vs Medals Graph Of Enterd Country

```python
def event_vs_medals_graph(country):
    nation_df = summer_athe_df[summer_athe_df['region'] == country]
    c= nation_df.groupby(['Event']).sum()[['Gold', 'Silver', 'Bronze']].reset_index()
    c['Total Medals'] = c['Gold'] + c['Silver'] + c['Bronze']
    plt.figure(figsize=(20,10))
    plt.plot(c['Event'], c['Total Medals'])
    plt.show()
    return plt.show()
```

```python
event_vs_medals_graph(country)
```

```python
def Sportwise_medals(country):
    nation_df = summer_athe_df[summer_athe_df['region'] == country]
    c= nation_df.groupby(['Sport']).sum()[['Gold', 'Silver', 'Bronze']].sort_values('Gold', ascending= False).reset_index()
    return c
```

```python
Sportwise_medals(country)
```
